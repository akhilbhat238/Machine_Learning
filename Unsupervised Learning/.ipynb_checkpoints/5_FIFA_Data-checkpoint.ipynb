```python
names = fifa_df['short_name'].tolist()
fifa_df = fifa_df.drop(['short_name'],axis=1)
```

```python
fifa_df.head()
```

## KMeans Clustering

```python
from sklearn import preprocessing
x=fifa_df.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
fifa_reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans = kmeans.fit(fifa_reduced)
labels = kmeans.predict(fifa_reduced)
centroid = kmeans.cluster_centers_
clusters = kmeans.labels_.tolist()
```

```python
fifa_reduced['cluster'] = clusters
fifa_reduced['name'] = names
fifa_reduced.columns = ['x','y','cluster','name']
```

```python
fifa_reduced.head(10)
```

## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='white')

ax = sns.lmplot(x='x',y='y',hue='cluster',data=fifa_reduced,size=15,fit_reg=False, scatter_kws={"s": 250},legend=False)

texts=[]
for x,y,s in zip(fifa_reduced.x,fifa_reduced.y,fifa_reduced.name):
    texts.append(plt.text(x,y,s))
    
ax.set(ylim=(-2,2))
plt.tick_params(labelsize=15)
plt.xlabel('PC1', fontsize=20)
plt.xlabel('PC2', fontsize=20)

plt.show()
```

## Hierarchical Clustering

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(10,14))

dendrogram = sch.dendrogram(sch.linkage(X_norm, method='average'),labels=names,leaf_font_size=13,orientation='right')
```

```python
plt.figure(figsize=(10,14))

dendrogram = sch.dendrogram(sch.linkage(X_norm, method='single'),labels=names,leaf_font_size=13,orientation='right')
```

```python
plt.figure(figsize=(10,14))

dendrogram = sch.dendrogram(sch.linkage(X_norm, method='centroid'),labels=names,leaf_font_size=13,orientation='right')
```

```python
plt.figure(figsize=(10,14))

dendrogram = sch.dendrogram(sch.linkage(X_norm, method='complete'),labels=names,leaf_font_size=13,orientation='right')
```
