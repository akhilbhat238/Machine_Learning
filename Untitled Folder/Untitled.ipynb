```python
players_summary = odi.groupby(['Player']).agg(
    Total_Runs=('Runs','sum'),
    Avg_Runs = ('Runs','mean'),
    Centuries = ('Runs',lambda v:(v>99).sum()),
    Fifties = ('Runs',lambda v:((v>49)&(v<100)).sum()),
    Ducks = ('Runs',lambda v:(v==0).sum()),
    Missed_Centuries = ('Runs',lambda v:((v>90)&(v<100)).sum())
)
```

```python
players_summary.head()
```
